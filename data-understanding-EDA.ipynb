# Summary of Data Understanding:

We have a relatively average sized dataset with 4920 row, with 133 parameters, 132 features and a target variable **'prognosis'** with 42 disease classifications that is completely balenced (120 instance of all the classes).
All the 132 features are 0/1 binary, one feature 'fluid_overload' is constant (all values at 0), will be dropped later.
We have more than 4000 rows duplicated, which means all these patients have the same disease and symptoms, will be feeding our models later two datasets, one with the duplicated and one without.
There are also 11 duplicated Columns, we will drop them.
Except for what was mentioned, the dataset has no other problems.
